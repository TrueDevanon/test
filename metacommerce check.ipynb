# используемые даты

```python
indexed = pd.read_sql('''select distinct indexed::date from novartis.parsing_nvs''',db.engine)
indexed = indexed.sort_values('indexed',ascending=False)
```

# сбор

```python
df_list = []
for i in tqdm(indexed['indexed'].astype('str').values):
    df = pd.read_sql(f'''
select * from novartis.parsing_nvs where indexed::date = '{i}'::date and gc_quality in ('number','exact') ''',db.engine)
    df.drop_duplicates(["catalog.name",
                                 "market.marketId",
                                 "market.marketName"], keep='last', inplace=True)
#     df = df.drop_duplicates(['catalog.name','net','point_geometry'],keep='last')
    df = pd.pivot_table(df, values='id', index='net', columns='catalog.name',
               aggfunc=pd.Series.nunique).fillna(0)
    df_list.append(df)
    
```

```python
def makeplot(name):
    dd = pd.DataFrame()
    for row,data in zip(df_list,indexed['indexed'].astype('str').to_list()):
        try:
            dd = pd.concat([dd,pd.DataFrame({data:row[name]}).T])
        except:
            print(f'Этот препарат не пришел с парсинга {data}')
    dd.iplot()
```
